```python
import numpy as np
import matplotlib.pyplot as plt
n = 1000

phi =np.random.exponential( 1, size=int(1e6) )
theta = np.random.exponential( 1, size=int(1e6) )
r = np.random.exponential( 1, size=int(1e6) )


index = ((phi**2)+(theta**2)+(r**2))<=1


phi = phi[index]
theta = theta[index]
r = r[index]
fig, axs = plt.subplots(3)
fig.suptitle('Vertically stacked subplots')
axs[0].hist(phi, bins=50, density=True)
axs[1].hist(theta, bins=50, density=True)
axs[2].hist(r, bins=50, density=True)

def f_x(phi,theta,r):
  x = r*np.sin(phi)*np.cos(theta)
  y = r*np.sin(phi)*np.sin(theta)
  z = r*np.cos(phi)
  det_J = np.sin(theta)
  return det_J*np.exp(((x**2)+(y**2)+(z**2))**0.5)


Integral = (2*(np.pi**2))*np.average( f_x(phi,theta,r))

Integral

```

```python
n = int(1e4)
x = np.random.rand(n+30)
def c_k(k,x):
  n = int(1e4)
  x_i = (x)[0:n]
  x_ik = (x)[k:(n+k)]
  return (1/n)*np.sum(x_i*x_ik)

k = np.linspace(1,30,30,dtype=int)
points = []
for i in k:
  points.append(c_k(i,x))
plt.style.use("seaborn-whitegrid")
plt.ylabel('c(K)')
plt.xlabel('K-esimo vecino')
plt.ylim((0.20,0.30))
plt.plot(k,points)


```
