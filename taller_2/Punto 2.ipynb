```python

import numpy as np
import matplotlib.pyplot as plt

xi,xf,N=-4,4,25
x=np.linspace(xi,xf,N)
y=x.copy()

X,Y = np.meshgrid(x,y)


V=2
R=2
def Potencial(x,y):
    phi=V*x*(1-(R**2/(x**2+y**2)))
    return phi


dx= lambda  x,y,f :  (f(x+h , y )-f(x-h , y ))/(2*h)

dy= lambda  x,y,f:  -(f(x,y+h )-f(x,y-h ))/(2*h)


# In[113]:


def campo(x,y):
    V_x=dx(x,y,Potencial)
    V_y=dy(x,y,Potencial)
    return V_x, V_y



u = 2
circle = plt.Circle((0, 0),2, color='r', fill=False)
fig=plt.figure(figsize=(5,5))
cilindro = (np.sqrt(X**2+Y**2))
ax=fig.add_subplot(1,1,1)
for i in range(0,25):
    for j in range(25):
      if (cilindro[i, j]) > 2:
        V_x[i,j],V_y[i,j] = campo(x[i],y[j])
        ax.quiver(x[i],y[j], V_x[i, j], V_y[i, j],color='blue',alpha=0.8)
ax.add_patch(circle) 
plt.show()

```
