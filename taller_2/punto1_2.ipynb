```python
import numpy as np
import matplotlib.pyplot as plt
#con h=0.05
xi, xf, h = -10., 10., 0.05
def f(x):
    return (1/(np.sqrt(1+np.e**(-(x**2)))))

def d_c(x,f,h):
    return (f(x+h)-f(x-h))/(2*h)

points = int((xf-xi)/h)

x = np.linspace(xi,xf,points)
y = f(x)
dev_c = d_c(x,f, h)
print("El error en cada uno de los puntos es",(dev_c/x))
plt.plot(x,y)
```

```python
#(Machine Learning analogy) 
M = [1,0,-1]
h= 0.05
def op_dc(M,h,x):
    sum = 0 
    for i in M:
        sum += i*f(x+i)
    return sum/(2*h)

x = np.linspace(xi,xf,points)
y = f(x)
dev_c = op_dc(M,h,x)
print()
plt.plot(x,dev_c)
```

```python
M_2 = [1,-2,1]
h= 0.05
def op_dc2(M,h,x):
    ind = -1
    sum = 0
    for i in M:
        sum += i*f(x+ind)
        ind+=1
    return sum/(4*(h**2))

points = int((xf-xi)/h)

x = np.linspace(xi,xf,points)
y = f(x)
dev_c2 = op_dc2(M_2,h,x)
plt.plot(x,dev_c2)
```
