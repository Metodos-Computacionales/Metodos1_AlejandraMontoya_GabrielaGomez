```python
def lanzamiento_monedas():
  set = [-1,1]
  frequency = 0
  n = int(1e5)
  for i in range(n): 
    select = random.choices(set, k=4)
    n_caras = select.count(1)
    if n_caras ==2:
      frequency +=1
  return round(frequency/i,3)
lanzamiento_monedas()
```
